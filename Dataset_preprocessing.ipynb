```python
import json

def create_text_rows(dataset_name):
    rows = []  # Initialize an empty list to store all text rows
    for i in range(len(dataset_name)):
        instruction = dataset_name.iloc[i]['Instruction']
        input_text = dataset_name.iloc[i]['Input']
        output = dataset_name.iloc[i]['Output']
        text_row = f"<s>[INST] {instruction} here are the inputs {input_text} [/INST] \\n {output} </s>"
        rows.append({
            "Text": text_row
        })
    return rows

def process_jsonl_file(output_file_path, dataset_name):
    rows = create_text_rows(dataset_name)  # Get all text rows
    with open(output_file_path, "w") as output_jsonl_file:
        for row in rows:
            output_jsonl_file.write(json.dumps(row) + "\n")

# Provide the path where you want to save the formatted dataset and the dataset_name for formatting data
process_jsonl_file("./wikiThai_testset_300rows.jsonl", df)
```
